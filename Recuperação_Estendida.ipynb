# Questão 1 (vale 1 ponto)
Suponha que você esteja projetando uma ponte suspensa e precise determinar a deformação máxima que ocorrerá em um determinado ponto da estrutura. A partir de análises e medições preliminares, você sabe que a deformação em um ponto na posição x da ponte pode ser aproximada por uma função $f(x) = 3x^2 - 2x + 5$. Precisamos saber a deformação quando $x$ se aproxima do valor $2$.

Utilizando a Fórmula das diferenças finitas encontre os limites para cinco pontos à esquerda do ponto desejado e cinco pontos à direita do ponto desejado. Depois traçe um gráfico desta função anote o limite no ponto desejado e imprima os limites nos cinco pontos à esquerda e à direita.




## Resposta

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 3 * x**2 - 2 * x + 5

def df_approx(x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

x = np.linspace(1, 3, 100)
y = f(x)

h = 0.1
x_left = [1.9, 1.8, 1.7, 1.6, 1.5]
x_right = [2.1, 2.2, 2.3, 2.4, 2.5]
y_left = [df_approx(x, h) for x in x_left]
y_right = [df_approx(x, h) for x in x_right]

plt.plot(x, y, label='f(x)')
plt.scatter(2, f(2), color='red', label='Ponto desejado (x=2)')
plt.scatter(x_left, y_left, color='green', label='Pontos à esquerda')
plt.scatter(x_right, y_right, color='blue', label='Pontos à direita')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

# Exercício de Limites e Descontinuidades

Considere as seguintes funções:

1. $$ f(x) = \tan(x) $$
2. $$ g(x) = \frac{1}{x-5} $$

Para cada função, trace o gráfico no intervalo de 0 a 10 usando Python, Matplotlib e NumPy. a)Determine, se existem, ou não descontinuidade para estas funções no intervalo solicitado. b)Caso exista alguma descontinuidade encontre o limite em um ponto qualquer à direita e outro a esquerda do ponto onde existe a descontinuidade. c)Você precisa plotar os gráficos destas funções usando python, e apenas as bibliotecas matplotlib, numpy e sympy.


## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.tan(x)

def df_approx(x, h):
    return (f(x + h) - f(x)) / h

x = np.linspace(0, 10, 1000)
y = f(x)

discontinuity_points = np.pi/2 + np.pi*np.arange(-3, 4)
for point in discontinuity_points:
    plt.axvline(x=point, color='red', linestyle='--')

h = 0.1
x_left = [np.pi/2 - 0.1, np.pi/2 - 0.2]
x_right = [np.pi/2 + 0.1, np.pi/2 + 0.2]
y_left = [df_approx(x, h) for x in x_left]
y_right = [df_approx(x, h) for x in x_right]

plt.plot(x, y)
plt.scatter(x_left, y_left, color='green', label='Limite à esquerda')
plt.scatter(x_right, y_right, color='blue', label='Limite à direita')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = tan(x) com descontinuidades')
plt.legend()
plt.grid(True)
plt.show()
```

#Questão 4 (vale 2 pontos)
Considere a função $f(x)= x^2 -1$. (a) Trace o gráfico desta função e, (b) usando a definição de derivada, encontre a inclinação da reta tangente nos pontos $(-1,0)$, $(0,-1)$ e $(2,3)$.

## Resposta Algébrica


$$
\begin{align*}
f(x) &= x^2 - 1 \\
\\
\text{Para o ponto } (-1, 0): \\
\\
\text{Passo 1:} \quad & f(-1) = (-1)^2 - 1 = 0 \\
\\
\text{Passo 2:} \quad & f(-1 + h) = (-1 + h)^2 - 1 = h^2 - 2h \\
\\
\text{Passo 3:} \quad & \lim_{{h \to 0}} \frac{{f(-1 + h) - f(-1)}}{h} = \lim_{{h \to 0}} \frac{{(h^2 - 2h) - 0}}{h} = \lim_{{h \to 0}} (h - 2) = -2 \\
\\
\text{A inclinação no ponto } (-1, 0) \text{ é } -2 \\
\\
\text{Para o ponto } (0, -1): \\
\\
\text{Passo 1:} \quad & f(0) = 0^2 - 1 = -1 \\
\\
\text{Passo 2:} \quad & f(0 + h) = (0 + h)^2 - 1 = h^2 - 1 \\
\\
\text{Passo 3:} \quad & \lim_{{h \to 0}} \frac{{f(0 + h) - f(0)}}{h} = \lim_{{h \to 0}} \frac{{(h^2 - 1) - (-1)}}{h} = \lim_{{h \to 0}} (h) = 0 \\
\\
\text{A inclinação no ponto } (0, -1) \text{ é } 0 \\
\\
\text{Para o ponto } (2, 3): \\
\\
\text{Passo 1:} \quad & f(2) = 2^2 - 1 = 3 \\
\\
\text{Passo 2:} \quad & f(2 + h) = (2 + h)^2 - 1 = h^2 + 4h + 3 \\
\\
\text{Passo 3:} \quad & \lim_{{h \to 0}} \frac{{f(2 + h) - f(2)}}{h} = \lim_{{h \to 0}} \frac{{(h^2 + 4h + 3) - 3}}{h} = \lim_{{h \to 0}} (h + 4) = 4 \\
\\
\text{A inclinação no ponto } (2, 3) \text{ é } 4 \\
\end{align*}
$$


## Resposta com código

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 1

x = np.linspace(-3, 3, 100)

y = f(x)

plt.plot(x, y, label='f(x) = x^2 - 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = x^2 - 1')
plt.grid(True)

# Ponto (-1, 0)
x1 = -1
h = 0.001
slope1 = (f(x1 + h) - f(x1)) / h

# Ponto (0, -1)
x2 = 0
slope2 = (f(x2 + h) - f(x2)) / h

# Ponto (2, 3)
x3 = 2
slope3 = (f(x3 + h) - f(x3)) / h

plt.scatter(x1, f(x1), color='red', label='Inclinação em (-1, 0): {:.2f}'.format(slope1))
plt.scatter(x2, f(x2), color='green', label='Inclinação em (0, -1): {:.2f}'.format(slope2))
plt.scatter(x3, f(x3), color='blue', label='Inclinação em (2, 3): {:.2f}'.format(slope3))

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

# Questão 5 (vale 1 ponto)
Você é um engenheiro de produção em uma fábrica de vidro que produz garrafas. A velocidade de produção das garrafas $V$ varia ao longo do tempo devido ao desgaste das máquinas e a fadiga dos trabalhadores. Depois de um mês de análise, você descobriu que a velocidade de produção é dada pela função $V(t) = 50 - 4t^2$ garrafas por hora, onde $t$ é o tempo em horas desde o início do turno de trabalho.
Seu gerente quer saber se há um momento específico durante o turno de trabalho em que a produção de garrafas é mais eficiente.
a) Como você poderia ajudar seu gerente a encontrar essa informação?
b) Se este momento existe, que momento é este?
c) Trace o gráfico da sua função usando apenas python, matplotlib, numpy e sympy e, caso este momento exista, marque este ponto no seu gráfico.

## Resposta  


## Para ajudar o gerente a encontrar o momento mais eficiente de produção de garrafas, podemos utilizar conceitos de cálculo diferencial para analisar a taxa de variação da velocidade de produção ao longo do tempo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

def V(t):
    return 50 - 4 * t**2

t = symbols('t')
V_derivative = diff(V(t), t)

coeffs = np.array([float(coeff) for coeff in V_derivative.as_coefficients_dict().values()])
critical_points = np.roots(coeffs)

t_vals = np.linspace(0, 5, 100)
V_vals = V(t_vals)
plt.plot(t_vals, V_vals)
plt.xlabel('Tempo (horas)')
plt.ylabel('Velocidade de Produção (garrafas por hora)')
plt.title('Gráfico da Velocidade de Produção')

if len(critical_points) > 0:
    V_max = max(V_vals)
    max_point = (t_vals[V_vals.argmax()], V_max)
    plt.scatter(max_point[0], max_point[1], color='red', label='Máximo')
    plt.annotate(f'Máximo: ({max_point[0]:.2f}, {max_point[1]:.2f})',
                 xy=max_point,
                 xytext=(max_point[0] + 0.5, max_point[1] - 5),
                 arrowprops=dict(arrowstyle='->', color='black'))

plt.grid(True)
plt.show()
```

# Questão 6 (vale 1 ponto)
Considere a função $f(x) = x^3 - 10x^2 + sin(x)$. Usando apenas o Python, numpy, sympy e matplotlib, trace o gráfico desta função e a reta tangente ao ponto $x=4$.

## Resposta

\begin{align*}
\text{Passo 1:} & \text{ Definir a função } f(x) = x^3 - 10x^2 + \sin(x). \\
\\
\text{Passo 2:} & \text{ Calcular a inclinação da reta tangente usando a derivada.} \\
f'(x) &= \frac{d}{dx} (x^3 - 10x^2 + \sin(x)) \\
f'(x) &= 3x^2 - 20x + \cos(x) \\
\\
\text{Passo 3:} & \text{ Calcular a inclinação da reta tangente no ponto } x = 4. \\
m &= f'(4) \\
m &= 3(4)^2 - 20(4) + \cos(4) \\
m &= -68 + \cos(4) \\
\\
\text{Passo 4:} & \text{ Calcular o valor de } f(x) \text{ no ponto } x = 4. \\
f(4) &= 4^3 - 10(4)^2 + \sin(4) \\
f(4) &= -96 + \sin(4) \\
\\
\text{Passo 5:} & \text{ Traçar a reta tangente ao ponto } x = 4 \text{ no gráfico de } f(x).
\end{align*}


```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 10*x**2 + np.sin(x)

x_value = 4

x = sp.symbols('x')
f_prime = sp.diff(x**3 - 10*x**2 + sp.sin(x), x)
slope = f_prime.subs(x, x_value)

y_value = f(x_value)

x = np.linspace(-10, 10, 100)

y = f(x)

plt.plot(x, y, label='f(x)')
plt.plot(x_value, y_value, marker='o', color='red', label=f'Tangente em x={x_value}')
plt.plot(x, slope*(x - x_value) + y_value, linestyle='--', color='blue', label='Reta Tangente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) e Reta Tangente')
plt.legend()
plt.grid(True)
plt.show()
```

#Questão 7 (vale 1 ponto)
Considere um fenômeno físico em que a posição de um objeto em movimento varia com o tempo. A taxa de variação da posição em relação ao tempo possui um componente trigonométrico e é dada pela função $f(t) = A * cos(t)$, onde $A$ é uma constante que determina a amplitude da variação. Sua tarefa é plotar o gráfico da função da taxa de variação e da função posição deste objeto considerando que $A$ é um número inteiro que varia entre 2 e 5.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

t = sp.Symbol('t')

def f_prime(A):
    return A * sp.cos(t)

def f(A):
    return sp.integrate(f_prime(A), t)

A_values = np.arange(2, 6)

plt.figure(figsize=(8, 6))
plt.title('Gráfico da taxa de variação e posição em função do tempo')

for i, A in enumerate(A_values):
    f_prime_values = [f_prime(A).subs(t, i) for i in np.linspace(0, 2*np.pi, 100)]
    f_values = [f(A).subs(t, i) for i in np.linspace(0, 2*np.pi, 100)]

    plt.plot(np.linspace(0, 2*np.pi, 100), f_prime_values, label=f'A = {A}')

    plt.plot(np.linspace(0, 2*np.pi, 100), f_values, linestyle='--')

    arrow_props = dict(facecolor='blue', arrowstyle='->')
    plt.annotate(f'A = {A}', xy=(2.5, f(A).subs(t, 2.5)), xytext=(3.5, f(A).subs(t, 2.5)),
                 color='blue', arrowprops=arrow_props)

plt.xlabel('t')
plt.ylabel('Taxa de variação / Posição')
plt.legend()
plt.grid(True)

plt.show()
```

#Questão 8 (vale 2 pontos)
Calcule a área entre as curvas $f(x) = -x^2$ e $g(x) = -x$, apenas no quarto quadrante do plano cartesiano, usando apenas álgebra e garantindo que todos os passos algébricos necessários estejam explicitados em notação matemática em uma célula de textos deste carderno de tarefas usando latex. Usando o python, matplotlib, numpy e sympy crie um gráfico destas funções sombreando a área entre elas em azul claro.

## Resposta



$$
\begin{align*}
\text{Passo 1:} & \text{ Definir as funções } f(x) \text{ e } g(x). \\
f(x) &= -x^2 \\
g(x) &= -x \\
\\
\text{Passo 2:} & \text{ Calcular os pontos de interseção entre as curvas.} \\
-f(x) &= -g(x) \\
-x^2 &= -x \\
x^2 - x &= 0 \\
x(x - 1) &= 0 \\
x = 0 \text{ ou } x &= 1 \\
\\
\text{Passo 3:} & \text{ Calcular a área entre as curvas.} \\
\text{Área} &= \int_{0}^{1} (g(x) - f(x)) \, dx \\
\text{Área} &= \int_{0}^{1} (-x - (-x^2)) \, dx \\
\text{Área} &= \int_{0}^{1} (-x + x^2) \, dx \\
\text{Área} &= \left[ -\frac{x^2}{2} + \frac{x^3}{3} \right]_{0}^{1} \\
\text{Área} &= -\frac{1}{2} + \frac{1}{3} \\
\text{Área} &= -\frac{1}{6}
\end{align*}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -x**2

def g(x):
    return -x

x_vals = np.linspace(0, 1, 100)

f_vals = f(x_vals)
g_vals = g(x_vals)

plt.plot(x_vals, f_vals, label='$f(x) = -x^2$')
plt.plot(x_vals, g_vals, label='$g(x) = -x$')

plt.fill_between(x_vals, f_vals, g_vals, where=(f_vals >= g_vals), color='lightgrey')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Gráfico das Funções $f(x) = -x^2$ e $g(x) = -x$')
plt.legend()

plt.grid(True)
plt.show()
```
