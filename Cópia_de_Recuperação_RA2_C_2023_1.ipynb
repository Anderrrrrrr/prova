# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

## 2.
$$h(x) = \frac{\ln(x)}{x}$$

```python
import numpy as np
import matplotlib.pyplot as plt


# Definindo a função 1
def f(x):
    return x**3 + 5*x**2 - 7*x + 9

# Definindo a derivada da função 1
def df(x):
    return 3*x**2 + 10*x - 7

# Gerando valores de x no intervalo desejado
x = np.linspace(-10, 10, 400)

# Calculando os valores de y para a função 1 e sua derivada
y = f(x)
dy = df(x)

# Plotando os gráficos
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^3 + 5x^2 - 7x + 9')
plt.plot(x, dy, label="f'(x) = 3x^2 + 10x - 7")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de f(x) e f\'(x)')
plt.legend()
plt.grid(True)
plt.show()



# Definindo a função 2
def h(x):
    return np.log(x) / x

# Definindo a derivada da função 2
def dh(x):
    return (1 - np.log(x)) / x**2

# Gerando valores de x no intervalo desejado
x = np.linspace(0.1, 10, 400)

# Calculando os valores de y para a função 2 e sua derivada
y = h(x)
dy = dh(x)

# Plotando os gráficos
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='h(x) = ln(x)/x')
plt.plot(x, dy, label="h'(x) = (1 - ln(x))/x^2")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de h(x) e h\'(x)')
plt.legend()
plt.grid(True)
plt.show()


```

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
import numpy as np

def f(x):
    return x**2 - 98*x + 4

x = np.linspace(0, 100, 1000)
y = f(x)

max_local = np.max(y[1:-1])
x_max_local = x[np.argmax(y[1:-1])+1]

print(f"O valor máximo local da função é {max_local} e ocorre em x = {x_max_local}")
```
