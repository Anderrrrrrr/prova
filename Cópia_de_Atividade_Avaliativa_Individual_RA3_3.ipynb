# Quest√£o 1 (vale at√© 2 pontos)
Determine a √°rea entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gr√°fico desta √°rea preenchendo a √°rea entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt

def y1(x):
    return x * np.exp(-x**2)

def y2(x):
    return x + 1

x = np.linspace(-2, 3, 1000)
area = np.minimum(y1(x), y2(x))

fig, ax = plt.subplots()
ax.plot(x, y1(x), label='y = x * exp(-x**2)', color='red')
ax.plot(x, y2(x), label='y = x + 1', color='green')
ax.fill_between(x, area, color='black', alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_title('√Årea entre as curvas')
ax.grid(True)

plt.show()
```

 grafico no codigo abaixo:

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a vari√°vel simb√≥lica
x = sp.Symbol('x')

# Definindo a fun√ß√£o f(x) = x^2 + 1
f = x**2 + 1

# Definindo a primitiva F(x) = x^3/3 + x
F = sp.integrate(f, x)

# Convertendo as express√µes simb√≥licas em fun√ß√µes num√©ricas
f_numeric = sp.lambdify(x, f)
F_numeric = sp.lambdify(x, F)

# Intervalo de x
x_vals = np.linspace(0, 2, 100)

# Avaliando as fun√ß√µes nos valores de x
f_vals = f_numeric(x_vals)
F_vals = F_numeric(x_vals)

# Plotando os gr√°ficos
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='f(x) = x^2 + 1', color='red')
plt.plot(x_vals, F_vals, label='F(x) = x^3/3 + x', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°ficos de f(x) e F(x)')
plt.axvline(x=0, color='gray', linestyle='--')
plt.axvline(x=2, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
