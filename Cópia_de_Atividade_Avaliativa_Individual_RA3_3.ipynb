# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt

def y1(x):
    return x * np.exp(-x**2)

def y2(x):
    return x + 1

x = np.linspace(-2, 3, 1000)
area = np.minimum(y1(x), y2(x))

fig, ax = plt.subplots()
ax.plot(x, y1(x), label='y = x * exp(-x**2)', color='red')
ax.plot(x, y2(x), label='y = x + 1', color='green')
ax.fill_between(x, area, color='black', alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_title('Área entre as curvas')
ax.grid(True)

plt.show()
```

a) para começar tem que encontrar a primitiva da função f(x) = x^2+1. Para isso, tem que encontrar uma função F(x) cuja derivada seja igual a f(x).A integral indefinida de f(x) é dada por:
∫(x^2+1)dx
Para calcular a integral indefinida, aplicamos as regras básicas de integração. Temos:
∫(x^2+1)dx=(x^3/3)+x+C
Onde C é uma constante de integração.
Agora, podemos calcular a integral definida da função f(x)= x^2 +1 no intervalo [0,2] substituindo os limites de integração na pimitiva que encontramos:

Agora, substituímos x = 2 e x = 0 na expressão acima:
((2^3/3)+2)-((0^3/3)+0)

8/3 + 2 - 0 = 8/3 +6/3 = 14/3

Portanto, o valor da integral definida:

 0


 grafico no codigo abaixo:

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definindo a variável simbólica
x = sp.Symbol('x')

# Definindo a função f(x) = x^2 + 1
f = x**2 + 1

# Definindo a primitiva F(x) = x^3/3 + x
F = sp.integrate(f, x)

# Convertendo as expressões simbólicas em funções numéricas
f_numeric = sp.lambdify(x, f)
F_numeric = sp.lambdify(x, F)

# Intervalo de x
x_vals = np.linspace(0, 2, 100)

# Avaliando as funções nos valores de x
f_vals = f_numeric(x_vals)
F_vals = F_numeric(x_vals)

# Plotando os gráficos
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='f(x) = x^2 + 1', color='red')
plt.plot(x_vals, F_vals, label='F(x) = x^3/3 + x', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráficos de f(x) e F(x)')
plt.axvline(x=0, color='gray', linestyle='--')
plt.axvline(x=2, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
