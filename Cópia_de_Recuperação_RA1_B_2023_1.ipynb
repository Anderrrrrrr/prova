# Questão 2
Considerando a função definida por: 
A função descontínua $f(x)$ é definida como:

$$
f(x) = 
\begin{cases} 
\sin(x) & \text{se } -10 \leq x < -3 \\
\cos(x) & \text{se } -3 \leq x < 0 \\
x^2 & \text{se } 0 \leq x < 3 \\
e^{x-10} & \text{se } 3 \leq x \leq 10
\end{cases}
$$

Trace o gráfico desta função, marque os pontos de descontinuidade, caso existam estes pontos, com um círculo vermelho, coloque o valor destes pontos. Não se esqueça de colocar uma legenda que permita o entendimento do seu gráfico indicando as curvas correspondentes a cada equação. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Piecewise, sin, cos, exp

# Definindo a função descontínua
x = symbols('x')
f = Piecewise((sin(x), x < -3), (cos(x), x < 0), (x**2, x < 3), (exp(x-10), x <= 10))

# Convertendo a função para uma função lambdify para plotagem
f_np = np.vectorize(lambda x: f.subs('x', x).evalf())
x_vals = np.linspace(-10, 10, 1000)
y_vals = f_np(x_vals)

# Encontrando os pontos de descontinuidade
discontinuity_points = [p for p in [-3, 0, 3] if f.subs('x', p).is_real is False]

# Plotando o gráfico
plt.plot(x_vals, y_vals, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função descontínua')
plt.legend()

# Marcando os pontos de descontinuidade com círculos vermelhos e adicionando as anotações
for p in discontinuity_points:
    plt.plot(p, f.subs('x', p).evalf(), 'ro')
    plt.annotate(f'({p}, {f.subs("x", p).evalf()})', xy=(p, f.subs('x', p).evalf()), xytext=(p, f.subs('x', p).evalf()+1),
                 arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.show()
```
